### <span style="color:blue">I.	Descripción del problema</span>

**El problema actual radica en que muchos corredores no tienen acceso a un análisis preciso de su tipo de pisada, lo que puede llevar a:**

- Técnicas de carrera ineficientes  
- Mayor riesgo de lesiones (fracturas por estrés, fascitis plantar, etc.)  
- Selección incorrecta de calzado deportivo  

**Según estudios recientes** (Fields et al., 2021; Napier et al., 2020), aproximadamente el **60-70% de las lesiones por correr** están relacionadas con patrones biomecánicos incorrectos. **Nuestra solución propone un sistema accesible con sensores de presión para identificar el tipo de pisada en tiempo real.**

### <span style="color:blue">II. Historia principal</span>

Sofia tiene 28 años y estudia mercadotecnia. Desde hace un par de años encontró en el running su escape del estrés. Corre tres veces por semana y sueña con participar en su primer medio maratón. Aunque ha mejorado su rendimiento, suele terminar con molestias en las rodillas y en el talón derecho. Siempre ha pensado que es parte del esfuerzo… hasta que un entrenador le comentó que podría estar corriendo de forma inadecuada.
Un día, mientras camina por el parque después de una carrera, escucha a otro corredor hablar sobre las tecnicas de correr y el uso de una suela para detectar tu pisada. No es una consulta médica ni una sesión técnica. Sofía decide investigar, sin expectativas.
Al principio no siente ningún cambio. Solo corre, como siempre. Pero con el paso de los días, empieza a notar algo diferente: sus zancadas se vuelven más estables, su pisada más ligera. Las molestias que la habían perseguido por meses comienzan a desaparecer. Ya no termina sus carreras con dolor, sino con una sonrisa.
Semanas después, al compartir su experiencia con otros corredores, se da cuenta de que ha aprendido a entender su cuerpo de una forma más íntima. Sabe cómo se mueve, qué debe ajustar, cuándo debe descansar. No fue magia, ni un gran cambio. Fue algo casi invisible, que la ayudó a descubrir lo que siempre estuvo ahí.
Ahora corre para disfrutar cada paso con conciencia y confianza.

# Requerimientos del Sistema de Análisis de Pisada

## 1. Identificación Automática de Pisada  
**Como corredor**, quiero que la suela detecte automáticamente qué parte del pie uso al correr, para saber si mi pisada es talonadora, de mediopié o de antepié.  

**Criterios de aceptación**:  
- *Dado* que el usuario está pisando con la suela puesta,  
- *Cuando* el sensor piezoeléctrico detecta presión en una zona específica,  
- *Entonces* el sistema determina el tipo de pisada y guarda esa información.  

---

## 2. Comunicación Clara del Resultado  
**Como corredor**, quiero recibir un mensaje que me diga claramente cuál es mi tipo de pisada y sus implicaciones, para entender mejor cómo corro y qué puedo mejorar.  

**Criterios de aceptación**:  
- *Dado* que la suela ya ha identificado el tipo de pisada,  
- *Cuando* el sistema procesa los datos,  
- *Entonces* se genera un mensaje con ventajas y desventajas de ese tipo de pisada que se imprime vía UART.  

---

## 3. Interacción sin Distracciones  
**Como corredor**, quiero que todo el análisis ocurra sin tener que usar el celular o detener mi entrenamiento, para enfocarme totalmente en correr.  

**Criterios de aceptación**:  
- *Dado* que el sistema está integrado en la suela,  
- *Cuando* estoy corriendo,  
- *Entonces* no necesito interactuar con una app ni hacer configuraciones.  

---

## 4. Seguimiento del Tipo de Apoyo  
**Como corredor**, quiero que la suela registre en qué zona del pie apoyo más, para detectar patrones o malos hábitos con el tiempo.  

**Criterios de aceptación**:  
- *Dado* que los sensores recogen datos de presión de cada zona,  
- *Cuando* termino una sesión,  
- *Entonces* puedo revisar qué zonas fueron más usadas: talón, mediopié o antepié.  

---

## 5. Experiencia Directa con Retroalimentación Básica  
**Como corredor**, quiero que la suela me dé una retroalimentación sencilla sin lenguaje técnico, para comprender mi pisada aunque no sea un experto.  

**Criterios de aceptación**:  
- *Dado* que el mensaje se genera por UART,  
- *Cuando* se imprime en consola,  
- *Entonces* debe mostrar un resultado claro y entendible: tipo de pisada y ventajas/desventajas.

### <span style="color:blue">Componentes de software</span>

El siguiente código implementa la lógica para la captura de pisadas con 6 sensores piezoeléctricos conectados a entradas ADC del STM32. Cada ciclo de muestreo evalúa si alguno de los sensores supera un umbral (200 unidades), indicando el inicio de una pisada. Durante el evento, se registra el valor máximo alcanzado por cada sensor. Cuando la pisada finaliza (todos los sensores por debajo del umbral), se almacena en un historial y se transmite por UART.

Esta lógica se repite hasta acumular 10 pisadas, tras lo cual se realiza un análisis para determinar la zona dominante (talón, mediopié o antepié), calcular porcentajes de distribución y enviar recomendaciones.

#### Prioridad: Normal (bucle principal, sin RTOS).
```
C
/* Dentro del bucle principal (main) */
while (1) {
  /* Leer valores de los 6 sensores */
  sensores[0] = Read_ADC(ADC_CHANNEL_0);
  sensores[1] = Read_ADC(ADC_CHANNEL_1);
  sensores[2] = Read_ADC(ADC_CHANNEL_4);
  sensores[3] = Read_ADC(ADC_CHANNEL_8);
  sensores[4] = Read_ADC(ADC_CHANNEL_11);
  sensores[5] = Read_ADC(ADC_CHANNEL_10);

  /* Detectar inicio de pisada */
  if (!evento_activo) {
    for (int i = 0; i < 6; i++) {
      if (sensores[i] > UMBRAL_PISADA) {
        zona_inicial = i;
        evento_activo = 1;
        break;
      }
    }
  }

  /* Capturar valores máximos mientras dure la pisada */
  if (evento_activo) {
    for (int i = 0; i < 6; i++) {
      if (sensores[i] > maximos[i]) maximos[i] = sensores[i];
    }

    /* Detectar fin de pisada */
    int activos = 0;
    for (int i = 0; i < 6; i++) {
      if (sensores[i] > UMBRAL_PISADA) activos++;
    }

    if (activos == 0) {
      /* Transmitir datos por UART */
      snprintf(msg, sizeof(msg), "[PISADA %d REGISTRADA]\r\nFuerzas: %lu, %lu, %lu, %lu, %lu, %lu\r\n", total_pisadas + 1, maximos[0], maximos[1], maximos[2], maximos[3], maximos[4], maximos[5]);
      HAL_UART_Transmit(&huart2, (uint8_t *)msg, strlen(msg), HAL_MAX_DELAY);

      /* Guardar historial */
      for (int i = 0; i < 6; i++) historial[total_pisadas][i] = maximos[i];
      total_pisadas++;

      /* Reiniciar */
      memset(maximos, 0, sizeof(maximos));
      zona_inicial = -1;
      evento_activo = 0;
    }
  }

  /* Analizar después de 10 pisadas */
  if (total_pisadas == MAX_PISADAS) {
    // Cálculo de promedios, zonas dominantes, porcentajes, riesgo y consejo
    // Se transmiten por UART
    total_pisadas = 0;
  }

  HAL_Delay(50);
}
```



#### Ejemplo: configurar los pines

La función Read_ADC configura el ADC para cada canal, inicia la conversión y devuelve el valor leído. Esta función se invoca para cada uno de los 6 sensores piezoeléctricos.


Prioridad: Normal (sin RTOS).
```
C
uint32_t Read_ADC(uint32_t channel) {
  ADC_ChannelConfTypeDef sConfig = {0};
  sConfig.Channel = channel;
  sConfig.Rank = 1;
  sConfig.SamplingTime = ADC_SAMPLETIME_3CYCLES;
  HAL_ADC_ConfigChannel(&hadc1, &sConfig);
  HAL_ADC_Start(&hadc1);
  HAL_ADC_PollForConversion(&hadc1, HAL_MAX_DELAY);
  return HAL_ADC_GetValue(&hadc1);
}
```


#### Transmisión UART

El sistema envía mensajes detallados de cada pisada, así como el análisis acumulado cada 10 pisadas. La transmisión se realiza usando la librería HAL y es síncrona.

Prioridad: Normal (sin RTOS).
```
/* Transmitir datos por UART */
HAL_UART_Transmit(&huart2, (uint8_t *)msg, strlen(msg), HAL_MAX_DELAY);
```
